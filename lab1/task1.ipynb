```python
import cv2

cv2.__version__
```

```python
img_path = 'data/one_piece.jpg'

img = cv2.imread(img_path)
img.shape
```

```python
import matplotlib.pyplot as plt

plt.imshow(img)
```

```python
img_grey = cv2.imread(img_path, 0)
plt.imshow(img_grey)
```

```python
b, g, r = img[100, 50]
b, g, r
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
plt.imshow(img_rgb[50:700, 500:1200])
```

```python
plt.imshow(cv2.resize(img_rgb, (300, 300)))
```

```python
import imutils

plt.imshow(imutils.resize(img_rgb, width=300))
```

```python
(h, w) = img_rgb.shape[:2]
center = (w // 2, h // 2)
M = cv2.getRotationMatrix2D(center, 45, 1.0)
rotated = cv2.warpAffine(img_rgb, M, (w, h))
plt.imshow(rotated)
```

```python
rotated = imutils.rotate(img_rgb, 45)
plt.imshow(rotated)
```

```python
blurred = cv2.GaussianBlur(img_rgb, (15, 15), 10)
plt.imshow(blurred)
```

```python
import numpy as np

comparison = np.hstack([img_rgb, blurred])
plt.imshow(comparison)
```

```python
output = img_rgb.copy()
cv2.rectangle(output, (700, 100), (1200, 500), (255, 255, 0), 6)
plt.imshow(output)
```

```python
output = np.zeros((300, 300, 3), dtype="uint8")
cv2.line(output, (0, 0), (300, 300), (255, 255, 0), 6)
plt.imshow(output)
```

```python
output = np.zeros((200, 200, 3), dtype="uint8")
points = np.array([[30, 30], [200, 10], [150, 200]], np.int32)
cv2.polylines(output, [points], True, (255, 255, 0), 3)
plt.imshow(output)
```

```python
output = np.zeros((200, 200, 3), dtype="uint8")
cv2.circle(output, (100, 100), 50, (255, 255, 0), 3)
plt.imshow(output)
```

```python
output = np.zeros((200, 200, 3), dtype="uint8")
cv2.putText(output, "Hello World", (10, 100), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 0), 3)
plt.imshow(output)
```
